```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.linspace(-3, 3, 100)
sine = np.sin(x_data)
cosine = np.cos(x_data)
tanh = np.tanh(x_data)
```

```python
%%mplfreeze trig_demo x_data sine cosine tanh
import lusca.render

with plt.style.context("lusca"):
    fig, axes = plt.subplots(
        1, 2, figsize=(7.0, 2.6), sharey=True, constrained_layout=True
    )

    axes[0].plot(x_data, sine, label="Sine")
    axes[0].plot(x_data, cosine, label="Cosine")
    axes[0].fill_between(x_data, sine - 0.1, sine + 0.1, alpha=0.2)
    axes[0].fill_between(x_data, cosine - 0.1, cosine + 0.1, alpha=0.2)

    axes[1].plot(sine, tanh, label="Tan vs Sine", linestyle="--")
    axes[1].plot(cosine, tanh, label="Tan vs Cosine", linestyle="--")

    axes[0].set_xlabel("X-axis")
    axes[0].set_ylabel("Y-axis")
    axes[1].set_xlabel("X2-axis")
    axes[1].set_ylabel("Y2-axis")

    axes[0].legend(loc="lower center", bbox_to_anchor=(0.5, -0.35), ncol=2)
    axes[1].legend(loc="lower center", bbox_to_anchor=(0.5, -0.35), ncol=2)
    plt.show()
```

```python
x_data = np.linspace(-3, 3, 200)
gaussian = np.exp(-(x_data**2))
exp_decay = np.exp(-np.abs(x_data))
```

```python
%%mplfreeze exp_demo x_data gaussian exp_decay
import lusca.render

with plt.style.context("lusca"):
    fig, ax = plt.subplots(figsize=(3.5, 2.6), constrained_layout=True)

    ax.plot(x_data, gaussian, label="Gaussian $e^{-x^2}$")
    ax.plot(x_data, exp_decay, label="Exponential $e^{-|x|}$")
    ax.fill_between(x_data, gaussian, exp_decay, alpha=0.2)

    ax.set_xlabel("X-axis")
    ax.set_ylabel("Function value")
    ax.legend(loc="upper right")
    plt.show()
```

```python
x_data = np.linspace(-2, 2, 400)
poly = x_data**3 - x_data
derivative = 3 * x_data**2 - 1
```

```python
%%mplfreeze poly_demo x_data poly derivative
import lusca.render

with plt.style.context("lusca"):
    fig, axes = plt.subplots(
        1, 2, figsize=(7.0, 2.6), sharey=False, constrained_layout=True
    )

    axes[0].plot(x_data, poly, label="$x^3 - x$")
    axes[0].axhline(0, color="k", linestyle=":", lw=1)
    axes[0].set_title("Cubic Polynomial")
    axes[0].set_xlabel("X-axis")
    axes[0].set_ylabel("Y-axis")
    axes[0].legend()

    axes[1].plot(x_data, derivative, color="tab:orange", label="Derivative $3x^2 - 1$")
    axes[1].axhline(0, color="k", linestyle=":", lw=1)
    axes[1].set_title("Derivative")
    axes[1].set_xlabel("X-axis")
    axes[1].legend()

    plt.show()
```
